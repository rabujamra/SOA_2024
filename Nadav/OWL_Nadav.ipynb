```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
a = 2
b = 5
c = 3
C0 = 2  # Cost for R0 (no treatment)
C1 = 4  # Cost for R1 (treatment)

# Define the functions R0 and R1
def R0(x):
    return a * x + (b - c)  # Baseline outcome (linear)

def R1(x):
    return 1.2 * R0(x) #a * x + b  # Improved outcome (linear)

# Define the ratios with cost for R0 and R1
def ratio_R0(x, k):
    return (R0(x)**k) / (C0**(1 - k))

def ratio_R1(x, k):
    return (R1(x)**k) / (C1**(1 - k))

# Define a range of k values
k_values = np.linspace(0, 1, 10)

# Assume we have 1000 cohorts, 100 in each decile
deciles = np.linspace(0, 10, 10)
cohort_size = 100

# Initialize arrays to store total R and C for each k
total_R = []
total_C = []

for k in k_values:
    R_sum = 0
    C_sum = 0
    
    for x in deciles:
        R0_val = ratio_R0(x, k)
        R1_val = ratio_R1(x, k)
        
        if R1_val > R0_val:
            # Choose to treat (use R1 and C1)
            R_sum += R1(x) * cohort_size
            C_sum += C1 * cohort_size
        else:
            # Choose not to treat (use R0 and C0)
            R_sum += R0(x) * cohort_size
            C_sum += C0 * cohort_size
    
    total_R.append(R_sum)
    total_C.append(C_sum)

# Calculate the final R/C ratio
final_ratios = np.array(total_R) / np.array(total_C)

# Plot the results
plt.figure(figsize=(10, 6))
#plt.plot(k_values, final_ratios, label='Total R/C Ratio', marker='o')
plt.plot(total_C, total_R, label='Total R/C Ratio', marker='o')

# Set axis properties
plt.xlabel('k values')
plt.ylabel('Total R')
plt.title('Total C')
plt.grid(True)
plt.legend()

# Show plot
plt.show()

# Print the results
for i, k in enumerate(k_values):
    print(f"For k = {k:.2f}, Total R = {total_R[i]:.2f}, Total C = {total_C[i]:.2f}, R/C = {final_ratios[i]:.2f}")
```
